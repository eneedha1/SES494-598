## Plotting

Plotting is not a part of core Python but is provided through one of several possible library modules. 

The most highly developed and widely used plotting package is MatPlotLib (<http://MatPlotLib.sourceforge.net/>). 

It is a powerful and flexible program that has become the standard for 2-d plotting with Python.

````python
import numpy as np
import matplotlib.pyplot as plt
````

There are other MatPlotLib sub-libraries, but the `pyplot` library provides nearly everything that you need for 2-d plotting. 

The standard prefix for it is `plt`. 


## Plot with `pyplot`

```python
plt.plot(x, y)
```

where `x` and `y` are arrays (or lists) that have the same size. 



Let's plot the sine function over the interval from 0 to $4\pi$. 


````python
import numpy as np
import matplotlib.pyplot as plt 


x = np.linspace(0, 4, 33) 
y = np.sin(x)
plt.plot(x, y);
````

```python
figure()
```  
creates a blank figure window. 

If it has no arguments, it creates a window that is 8 inches wide and 6 inches high by default.

```python
figure()
```  
You can create a window whose size differs from the default using the optional keyword argument `figsize`, as we have done here. 

If you use `figsize`, set it equal to a 2-element tuple where the elements are the width and height, respectively, of the plot. 

```python
plot(x, y,` *optional arguments* `)
```

Graphs the $x$-$y$ data in the arrays `x` and `y`.

The third argument is a format string that specifies the color and the type of line or symbol that is used to plot the data. 

```python
plot(x, y,` *optional arguments* `)
```

The string `'ro'` specifies a red (`r`) circle (`o`). The string `'b-'` specifies a blue (`b`) solid line (`-`). The keyword argument `label` is set equal to a string that labels the data if the `legend` function is called subsequently.


```python
xlabel(` *string* `)```

takes a string argument that specifies the label for the graph's $x$-axis.

```python
ylabel(` *string* `)
```  
takes a string argument that specifies the label for the graph's $y$-axis.

```python
legend()
```  
makes a legend for the data plotted. Each $x$-$y$ data set is labeled using the string that was supplied by the `label` keyword in the `plot` function that graphed the data set. The `loc` keyword argument specifies the location of the legend.

```python
savefig(` *string* `)
```  
saves the figure to data data file with a name specified by the string argument. The string argument can also contain path information if you want to save the file so some place other than the default directory.

```python
show()
```  
displays the plot on the computer screen. No screen output is produced before this function is called.

## Error bars


````python
xdata, ydata, yerror = np.loadtxt('./data/expDecayData.txt', unpack=True)
x = np.linspace(0, 45, 128) # create theoretical fitting curve
y = 1.1+ 3.0*x*np.exp(-(x/10.0)**2)
````

````python
plt.figure(1, figsize = (6,4) )
plt.plot(x, y, 'b-', label="theory")
plt.errorbar(xdata, ydata, fmt='ro', label="data",
             xerr=0.75, yerr=yerror, ecolor='black')
plt.xlabel('x')
plt.ylabel('transverse displacement')
plt.legend(loc='upper right')
plt.savefig('./data/ExpDecay.pdf')
````

```python
plt.errorbar(xdata, ydata, fmt='ro', label="data",
             xerr=0.75, yerr=yerror, ecolor='black')
```

When error bars are desired, you replace the `plot` function with the `errorbar` function. 

The first two arguments of the `errorbar` function are the `x` and `y` arrays to be plotted. 

```python
plt.errorbar(xdata, ydata, fmt='ro', label="data",
             xerr=0.75, yerr=yerror, ecolor='black')
```

The keyword `fmt` *must be used* to specify the format of the points to be plotted; the format specifiers are the same as for `plot`. 

The keywords `xerr` and `yerr` are used to specify the $x$ and $y$ error bars. 

The color of the error bars is set with the keyword `ecolor`.

### Subplots

Often you want to create two or more graphs and place them next to one another, generally because they are related to each other in some way. 

````python
theta = np.arange(0.01, 8., 0.04)
ysin = np.sin(theta)
ycos = np.cos(theta)
````

````python
plt.figure(1)
plt.subplot(2, 3, 1)
plt.plot(theta, ysin)
plt.subplot(2, 3, 5)
plt.plot(theta, ycos)
plt.subplot(2, 3, 3)
plt.plot(theta, ycos)
````

The function `subplot` creates the two subplots in the above figure. 

`subplot` has three arguments. 

The first specifies the number of rows that the figure space is to be divided into, it's two. 

The second specifies the number of columns that the figure space is to be divided into, it's one. 

The third argument specifies which rectangle the will contain the plot specified by the following function calls. 

```plt.subplot(2, 1, 2)``` specifies that the plotting commands that follow will be act on the first box. 

## Semi-log plots

For data sets that vary exponentially in the independent variable, it is often useful to use one or more logarithmic axes. 

MatPlotLib provides two functions for making semi-logarithmic plots, `semilogx` and `semilogy`, for creating plots with logarithmic $x$ and $y$ axes, with linear $y$ and $x$ axes, respectively. We illustrate their use in the program below, which made the above plots.

````python
time, counts, unc = np.loadtxt('./data/SemilogDemo.txt', unpack=True)
# create theoretical fitting curve
tau = 20.2 # Phosphorus-32 half life = 14 days; tau = t_half/ln(2)
N0 = 8200. # Initial count rate (per second)
t = np.linspace(0, 180, 128)
N = N0 * np.exp(-t/tau)
````

````python
plt.figure(1, figsize = (10,4) )
plt.subplot(1, 2, 1)
plt.plot(t, N, 'b-', label="theory"); plt.plot(time, counts, 'ro', label="data")
plt.xlabel('time (days)'); plt.ylabel('counts per second')
plt.legend(loc='upper right')
plt.subplot(1, 2, 2)
plt.semilogy(t, N, 'b-', label="theory"); plt.semilogy(time, counts, 'ro', label="data")
plt.xlabel('time (days)'); plt.ylabel('counts per second')
plt.legend(loc='upper right')
plt.tight_layout()
````

The `semilogx` and `semilogy` functions work the same way as the `plot` function. You just use one or the other depending on which axis you want to be logarithmic.



## Log-log plots

MatPlotLib can also make log-log or double-logarithmic plots using the function `loglog`. It is useful when both the $x$ and $y$ data span many orders of magnitude. Data that are described by a power law $y=Ax^b$, where $A$ and $b$ are constants, appear as straight lines when plotted on a log-log plot. Again, the `loglog` function works just like the `plot` function but with logarithmic axes.
